## Trefethen & Bau — Numerical Linear Algebra

Python demonstrations using NumPy and Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

## Vector Norms

The three main $p$-norms for $x = (3, -4, 1)^T$.

```python
x = np.array([3, -4, 1], dtype=float)

print(f"x = {x}")
print(f"||x||_1   = {np.linalg.norm(x, 1)}")
print(f"||x||_2   = {np.linalg.norm(x, 2):.4f}")
print(f"||x||_inf = {np.linalg.norm(x, np.inf)}")
```

## Contour Lines of Vector Norms (Unit Balls)

The unit ball $\{x : \|x\| \leq 1\}$ in $\mathbb{R}^2$ has a different shape for each norm:
- **1-norm:** diamond
- **2-norm:** circle
- **$\infty$-norm:** square

```python
x1 = np.linspace(-1.2, 1.2, 300)
x2 = np.linspace(-1.2, 1.2, 300)
X1, X2 = np.meshgrid(x1, x2)

norm_2   = np.sqrt(X1**2 + X2**2)
norm_1   = np.abs(X1) + np.abs(X2)
norm_inf = np.maximum(np.abs(X1), np.abs(X2))

fig, axes = plt.subplots(1, 3, figsize=(14, 4))

for ax, Z, title in zip(axes,
                         [norm_1, norm_2, norm_inf],
                         ['1-norm', '2-norm', r'$\infty$-norm']):
    ax.contour(X1, X2, Z, levels=np.arange(0.2, 1.6, 0.2), cmap='viridis')
    ax.set_aspect('equal')
    ax.set_title(title, fontsize=13)
    ax.set_xlabel('$x_1$')
    ax.set_ylabel('$x_2$')
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## Induced Matrix Norms: The 2-Norm

$\|A\|_2 = \max_{\|x\|_2=1} \|Ax\|_2$

Map the unit circle through $A$ and see how much it stretches.

```python
A = np.array([[-2, 0],
              [-1, 3]], dtype=float)

t = np.linspace(0, 2 * np.pi, 300)
X = np.array([np.cos(t), np.sin(t)])  # unit circle
AX = A @ X

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[0].scatter(X[0], X[1], c=t, cmap='hsv', s=4)
axes[0].set_aspect('equal')
axes[0].set_title('x  (unit circle)')
axes[0].grid(True, alpha=0.3)

axes[1].scatter(AX[0], AX[1], c=t, cmap='hsv', s=4)
axes[1].set_aspect('equal')
axes[1].set_title('Ax  (image under A)')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# ||Ax||_2 as a function of angle, compared to ||A||_2
norms_Ax = np.sqrt(np.sum(AX**2, axis=0))
norm_A_2 = np.linalg.norm(A, 2)

plt.figure(figsize=(7, 3.5))
plt.plot(t, norms_Ax, label=r'$\|Ax\|_2$')
plt.axhline(norm_A_2, color='r', linestyle=':', label=rf'$\|A\|_2 = {norm_A_2:.4f}$')
plt.xlim(0, 2 * np.pi)
plt.xlabel('angle')
plt.ylabel(r'$\|Ax\|_2$')
plt.title('2-norm of Ax over the unit circle')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## Induced Matrix Norms: The $\infty$-Norm

$\|A\|_\infty = \max_{\|x\|_\infty=1} \|Ax\|_\infty$

Map the $\infty$-norm unit ball (a square) through $A$.

```python
# Points on the boundary of the inf-norm unit ball (a square)
s = np.linspace(-1, 1, 100)
o = np.ones_like(s)
x1_sq = np.concatenate([s,  o,  s, -o])
x2_sq = np.concatenate([o,  s, -o,  s])
X_sq = np.array([x1_sq, x2_sq])
AX_sq = A @ X_sq

colors = np.arange(len(x1_sq))

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[0].scatter(X_sq[0], X_sq[1], c=colors, cmap='hsv', s=4)
axes[0].set_aspect('equal')
axes[0].set_title(r'x  ($\infty$-norm unit ball)')
axes[0].grid(True, alpha=0.3)

axes[1].scatter(AX_sq[0], AX_sq[1], c=colors, cmap='hsv', s=4)
axes[1].set_aspect('equal')
axes[1].set_title(r'Ax  (image under A)')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# ||Ax||_inf over the square boundary
norms_Ax_inf = np.max(np.abs(AX_sq), axis=0)
norm_A_inf = np.linalg.norm(A, np.inf)

plt.figure(figsize=(7, 3.5))
plt.plot(norms_Ax_inf, label=r'$\|Ax\|_\infty$')
plt.axhline(norm_A_inf, color='r', linestyle=':', label=rf'$\|A\|_\infty = {norm_A_inf}$')
plt.ylabel(r'$\|Ax\|_\infty$')
plt.title(r'$\infty$-norm of Ax over the $\infty$-norm unit ball boundary')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## Computing 1-Norm and $\infty$-Norm Directly

- $\|A\|_1$ = max absolute column sum
- $\|A\|_\infty$ = max absolute row sum

```python
print("A =")
print(A)

onenorm_manual = np.max(np.sum(np.abs(A), axis=0))
infnorm_manual = np.max(np.sum(np.abs(A), axis=1))

print(f"\n||A||_1   = {onenorm_manual}  (max abs column sum)")
print(f"||A||_inf = {infnorm_manual}  (max abs row sum)")
print(f"||A||_2   = {np.linalg.norm(A, 2):.4f}  (largest singular value)")

# Verify against numpy
print(f"\nnp.linalg.norm(A, 1)      = {np.linalg.norm(A, 1)}")
print(f"np.linalg.norm(A, np.inf) = {np.linalg.norm(A, np.inf)}")
```

## Unitary Matrices Preserve the 2-Norm

A unitary (orthogonal) matrix preserves Euclidean distances — it's a rigid rotation/reflection.

Here we rotate points on a partial sphere in 3D.

```python
# Generate points on a partial sphere
phi = np.linspace(0, 1.5 * np.pi, 60)
theta = np.linspace(0.25 * np.pi, np.pi, 50)
Phi, Theta = np.meshgrid(phi, theta)
Phi = Phi.ravel()
Theta = Theta.ravel()

x_pts = np.column_stack([
    np.cos(Phi) * np.sin(Theta),
    np.sin(Phi) * np.sin(Theta),
    np.cos(Theta)
])

# Create an orthogonal matrix from QR of a fixed matrix
magic3 = np.array([[8, 1, 6], [3, 5, 7], [4, 9, 2]], dtype=float)
Q, _ = np.linalg.qr(magic3)

Qx_pts = (Q @ x_pts.T).T
colr = np.sum(x_pts, axis=1)

fig = plt.figure(figsize=(12, 5))

ax1 = fig.add_subplot(121, projection='3d')
ax1.scatter(x_pts[:, 0], x_pts[:, 1], x_pts[:, 2], c=colr, s=8, cmap='coolwarm')
ax1.set_title('x (partial sphere)')
ax1.set_box_aspect([1, 1, 1])

ax2 = fig.add_subplot(122, projection='3d')
ax2.scatter(Qx_pts[:, 0], Qx_pts[:, 1], Qx_pts[:, 2], c=colr, s=8, cmap='coolwarm')
ax2.set_title('Qx (rotated sphere — same shape)')
ax2.set_box_aspect([1, 1, 1])

plt.tight_layout()
plt.show()
```

```python
# Verify: norms are preserved
orig_norms = np.linalg.norm(x_pts, axis=1)
rot_norms = np.linalg.norm(Qx_pts, axis=1)

print(f"Max difference in norms: {np.max(np.abs(orig_norms - rot_norms)):.2e}")
print(f"All points have ||x|| = 1: {np.allclose(orig_norms, 1.0)}")
```

## Equivalence of Norms

$\|x\|_\infty \leq \|x\|_2 \leq \|x\|_1$ always holds, with worst-case gaps of $\sqrt{n}$ and $n$.

```python
x = np.array([1.0, 1.0, 1.0])
n = len(x)

n1 = np.linalg.norm(x, 1)
n2 = np.linalg.norm(x, 2)
ninf = np.linalg.norm(x, np.inf)

print(f"x = {x},  n = {n}")
print(f"||x||_inf = {ninf}")
print(f"||x||_2   = {n2:.4f}  (ratio to inf-norm: {n2/ninf:.4f}, sqrt(n) = {np.sqrt(n):.4f})")
print(f"||x||_1   = {n1}      (ratio to inf-norm: {n1/ninf:.4f}, n = {n})")
```

## The Crucial Inequality: $\|Ax\| \leq \|A\| \, \|x\|$

Verify numerically for many random vectors.

```python
A = np.array([[-2, 0],
              [-1, 3]], dtype=float)
norm_A = np.linalg.norm(A, 2)

np.random.seed(42)
ratios = []
for _ in range(10000):
    x = np.random.randn(2)
    ratio = np.linalg.norm(A @ x) / np.linalg.norm(x)
    ratios.append(ratio)

plt.figure(figsize=(7, 3.5))
plt.hist(ratios, bins=80, density=True, alpha=0.7, label=r'$\|Ax\|/\|x\|$')
plt.axvline(norm_A, color='r', linestyle='--', linewidth=2,
            label=rf'$\|A\|_2 = {norm_A:.4f}$')
plt.xlabel(r'$\|Ax\|_2 / \|x\|_2$')
plt.ylabel('Density')
plt.title(r'$\|Ax\|/\|x\|$ is always $\leq \|A\|$')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## Submultiplicativity: $\|AB\| \leq \|A\| \, \|B\|$

```python
np.random.seed(7)
A = np.random.randn(3, 3)
B = np.random.randn(3, 3)

norm_AB = np.linalg.norm(A @ B, 2)
bound = np.linalg.norm(A, 2) * np.linalg.norm(B, 2)

print(f"||A||   = {np.linalg.norm(A, 2):.4f}")
print(f"||B||   = {np.linalg.norm(B, 2):.4f}")
print(f"||AB||  = {norm_AB:.4f}")
print(f"||A||*||B|| = {bound:.4f}")
print(f"||AB|| <= ||A||*||B||: {norm_AB <= bound + 1e-14}")
```
